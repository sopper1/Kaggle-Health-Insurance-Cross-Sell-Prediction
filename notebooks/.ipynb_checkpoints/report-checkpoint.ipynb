## Analysis
### Functions for later use
Unfortunately, I don't know how to define functions and use them from a .py file outside of a notebook yet.

```python
def get_response_chance(data_subset):
    positive = data_subset[data_subset['Response'] == "Yes"]
    try:
        return len(positive)/len(data_subset)
    except:
        return 0
```

Response rate refers to the percentage of samples that have a response of 1, and would be interested in vehicle insurance. It is essentially the mean response over a given set. This is preferrend over the median since the meadin would only give us a 0 or a 1, which makes the graphs a lot less meaningful.

```python
print("Total response rate:", get_response_chance(insurance_data))
```

```python
def get_response_continuous(data, column, bin_count = 20, axis = None):

    column_min, column_max = min(data[column]), max(data[column])

    step = int((column_max - column_min) / bin_count)
    bins = range(int(column_min), int(column_max - step), step)
    response_chart = []

    for i in bins:
        temp_min, temp_max = i, i + step
        column_subset = data[data[column] >= temp_min]
        column_subset = column_subset[column_subset[column] < temp_max]
        response_chart.append((i, get_response_chance(column_subset)))

    dataFrame = pd.DataFrame(data=response_chart, columns=[column, 'Response chance'])
    
    if axis == None:
        dataFrame.plot.line(x = column, y = 'Response chance', title = 'Response Chance Based on ' + column)
    else:
        dataFrame.plot.line(x = column, y = 'Response chance', title = 'Response Chance Based on ' + column, ax = axis)
    
def get_response_categorical(data, column, axis = None):
    
    bins = data[column].unique()
    response_chart = []

    for i in bins:
        column_subset = insurance_data[insurance_data[column] == i]
        response_chart.append((i, get_response_chance(column_subset)))

    dataFrame = pd.DataFrame(data=response_chart, columns=[column, 'Response chance'])
    
    if axis == None:
        dataFrame.sort_values(column).plot.bar(x = column, y = 'Response chance', title = 'Response Chance Based on ' + column)
    else:
        dataFrame.sort_values(column).plot.bar(x = column, y = 'Response chance', title = 'Response Chance Based on ' + column, ax = axis)
```

### Graphs
The following graphs all show the effect of one variable in the trainset on the chances of a person responding (showing interest in the company's vehicle insurance). Correlations appear between Age, Vehicle_Age, Vehicle_Damage, and Previously_Insured. I left out the id variable since that should be independent of response. Drivers_License should perhaps be ignored since over 99% of the samples had a driver's license.

```python
fig, ax = plt.subplots(nrows = 5, ncols = 2, figsize = (20, 20))
fig.tight_layout(pad = 8.0)

quantative_cols = ['Age', 'Annual_Premium', 'Policy_Sales_Channel', 'Vintage']
qualitative_cols = ['Gender', 'Driving_License', 'Region_Code', 'Previously_Insured', 'Vehicle_Age', 'Vehicle_Damage']

i = 0

for col in quantative_cols:
    get_response_continuous(insurance_data, col, axis = ax[int(i/2)][i%2])
    i += 1
    
for col in qualitative_cols:
    get_response_categorical(insurance_data, col, axis = ax[int(i/2)][i%2])
    i += 1
```

## Brief Data Summary

Many categorical features heavily affect the response rate when treated independently:
   * If a customer was previously insured, then the response rate is almost 0%
   * For every year the the customer has owned the vehicle up until 2 years, the response rate goes up by about 15%
   * If the customer's vehicle was never damaged, then the reponse rate is near 0%
   
Age is the continuous feature most strongly correlated with response rate. The response rate would peak between age 30 and 45 at about 20%, then falls by about 1% every three years.
